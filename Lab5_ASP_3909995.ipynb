* **Series:**

The primary data type we will explore in pandas is the Series data type, which bears a striking resemblance to a NumPy array. In fact, a Series is constructed on top of the NumPy array object. The key distinction between a NumPy array and a Series is that a Series can possess axis labels, allowing for indexing based on labels rather than just numerical positions **[4]**.

A Series can be created by converting a list, NumPy array, or dictionary into one.

```python
labels = ['a','b','c']
mylist = [10,20,30]
arr = np.array([10,20,30])
d = {'a':10,'b':20,'c':30}
```

```python
pd.Series(data=mylist)
```

The given code statement `pd.Series(data=mylist)` creates a Series object in pandas using the `pd.Series()` constructor function. The `data` parameter is set to `mylist`, which is a list containing the values `[10, 20, 30]`. This creates a Series where the values from `mylist` are assigned default numerical indices `[0, 1, 2]`. If the `labels` list is available and matches the length of `mylist`, it can be provided as the `index` parameter to assign custom labels to the Series **[4]**.

```python
arr = np.array([10,20,30])
arr
```

```python
print(arr)
```

The variable `arr` represents a NumPy array containing the values `[10, 20, 30]`. It is created using the `np.array()` function from the NumPy library. This array can be used for various numerical operations and array manipulations provided by NumPy **[4]**.

```python
pd.Series(data=mylist, index=labels)
```

The code statement `pd.Series(data=mylist, index=labels)` creates a Series object in pandas using the `pd.Series()` constructor function. The `data` parameter is set to `mylist`, which is a list containing the values `[10, 20, 30]`. The `index` parameter is set to `labels`, which is a list containing the labels `['a', 'b', 'c']`. This creates a Series where the values from `mylist` are associated with the corresponding labels from `labels`. The resulting Series will have custom labels `['a', 'b', 'c']` instead of the default numerical indices **[4]**.

```python
pd.Series(arr,labels)
```

The code statement `pd.Series(arr, labels)` creates a Series object in pandas using the `pd.Series()` constructor function. The `arr` parameter is a NumPy array `[10, 20, 30]`, and the `labels` parameter is a list `['a', 'b', 'c']`. This creates a Series where the values from the NumPy array `arr` are paired with the corresponding labels from the `labels` list. The resulting Series would have the values `[10, 20, 30]` and the labels `['a', 'b', 'c']` **[4]**.

```python
pd.Series(data=labels)
```

The code statement `pd.Series(data=labels)` creates a Series object in pandas using the `pd.Series()` constructor function. Here, the `data` parameter is set to `labels`, which is a list containing the values `['a', 'b', 'c']`. This creates a Series where the values from `labels` are assigned default numerical indices `[0, 1, 2]`.
```python
pd.Series([sum,print,len])
```

The code statement `pd.Series([sum, print, len])` creates a Series object in pandas. Here, the list `[sum, print, len]` contains built-in Python functions as its elements. By passing this list to the `pd.Series()` constructor function, a Series is created where the functions `sum`, `print`, and `len` become the values of the Series **[4]**.

Since functions are objects in Python, they can be stored and manipulated just like any other data type. In this case, the Series is created with the default numerical indices `[0, 1, 2]`. The functions `sum`, `print`, and `len` can be accessed from the Series using these indices **[4]**.

* **Using as Index:**

The key to using a Series lies in understanding its index. The index names or numbers are utilized by pandas for quick information lookup, resembling a hash table or dictionary **[5]**.

Examples of information retrieval from a Series will be demonstrated. Two Series, namely `sales_Q1` and `sales_Q2`, will be created.

```python
sales_Q1 = pd.Series(data=[250,450,200,150],index = ['USA', 'China','India', 'Brazil'])
sales_Q1
```

```python
sales_Q2 = pd.Series([260,500,210,100],index = ['USA', 'China','India', 'Japan'])
sales_Q2
```

```python
sales_Q1['USA']
```

The statement `sales_Q1['USA']` is accessing a specific element in the `sales_Q1` Series. By using the index label 'USA', it retrieves the corresponding value associated with that label. In this case, it would return the sales value for the USA in the first quarter, which is 250.

```python
sales_Q2['Japan']
```

The statement `sales_Q2['Japan']` is accessing a specific element in the `sales_Q2` Series. By using the index label 'Japan', it retrieves the corresponding value associated with that label. In this case, it would return the sales value for Japan in the second quarter, which is 100.

```python
sales_Q1 + sales_Q2
```

The expression `sales_Q1 + sales_Q2` performs element-wise addition between the `sales_Q1` and `sales_Q2` Series. When performing this operation, the values with matching index labels are added together. For example, the sales values for 'USA' in both `sales_Q1` and `sales_Q2` are added, as well as the sales values for 'China', 'India', and 'Brazil' that have corresponding index labels in both Series. The resulting Series will contain the sum of the corresponding sales values for each index label **[5]**.

* **Data frames:**

Data Frames are considered the primary workhorse of pandas, serving as a crucial component **[6]**. A DataFrame can be conceptualized as a collection of Series objects that share the same index. It is a 2-dimensional data structure, resembling a 2-dimensional array or a table consisting of rows and columns **[6]**.

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df)
```

The code snippet above demonstrates the creation of a DataFrame object using the provided data. The data is structured as a dictionary, where the keys represent the column names and the values are lists containing the corresponding data for each column. In this case, the DataFrame has two columns: "calories" and "duration". 

By calling `pd.DataFrame(data)`, the data dictionary is passed as an argument to the `DataFrame()` function from the pandas library, which converts it into a tabular data structure. The resulting DataFrame, assigned to the variable `df`, has the column names as its headers and the values from the lists as its rows **[6]**. Finally, `print(df)` is used to display the DataFrame in a readable format, showing the values for each column along with their respective row indices **[6]**.

* **Removing Column/Row:**

The drop function in pandas is used to remove specified rows or columns from a DataFrame. It takes labels or positions as arguments and returns a new DataFrame with the specified elements removed. The original DataFrame is not modified unless the inplace parameter is set to True **[8]**.

```python
df = df.drop('new',axis =1)
df
```

The column named 'new' is removed from the DataFrame `df` using the code `df = df.drop('new', axis=1)`. The `drop()` method is employed to eliminate specific columns or rows from a DataFrame. In this case, the 'new' column is specified as the argument to `drop()`, with the additional parameter `axis=1` indicating the operation should be performed along the columns **[8]**. As a result, the DataFrame `df` is modified, and the 'new' column is no longer present. The updated DataFrame is then assigned back to the variable `df`, effectively replacing the original DataFrame.

The code `df.drop('C', axis=0)` removes the row labeled 'C' from the DataFrame `df`. It returns a new DataFrame with the specified row eliminated. The original DataFrame remains unchanged unless the result is assigned back to `df` or the `inplace` parameter is set to `True`.

* **Working with rows:**

The `loc` function in pandas is a label-based indexer used for selection and indexing of rows and columns in a DataFrame. It allows you to retrieve data based on specific index labels, enabling precise and explicit selection **[9]**.

Both `loc` and `iloc` functions enable efficient and convenient access to specific data elements or subsets within a DataFrame, enhancing the capabilities of data manipulation and analysis in pandas **[9]**.

```python
df.loc['A']
```

`df.iloc[0]` retrieves the row at the specified position 0 (the first row) from the DataFrame `df`. It returns a Series containing the values in the first row.

`df.iloc[0:2]` selects a range of rows from the DataFrame `df` based on their positions. It returns a new DataFrame containing the rows at positions 0 and 1 (the first two rows).

* **Missing data:**

Missing data, also referred to as NaN (Not a Number) or null values, represents the absence or unknown values in a dataset. In pandas, missing data is commonly represented by NaN **[11]**.


```python
df = pd.DataFrame({'A':[1,2,np.nan,4],
'B':[5,np.nan,np.nan,8],
'C':[10,20,30,40]})
df
```

`df = pd.DataFrame({'A':[1,2,np.nan,4], 'B':[5,np.nan,np.nan,8], 'C':[10,20,30,40]})` creates a DataFrame named `df` with three columns ('A', 'B', 'C') and four rows. The DataFrame contains missing values (NaN) in columns 'A' and 'B'.

```python
df.dropna()
```

`df.dropna()` drops any rows containing missing values (NaN) from the DataFrame `df`. This operation removes rows with NaN values, resulting in a new DataFrame with only the rows that have complete data.

```python
df.dropna(axis=1)
```

`df.dropna(axis=1)` drops any columns containing missing values (NaN) from the DataFrame `df`. This operation removes columns with NaN values, resulting in a new DataFrame that retains only the columns with complete data.

```python
df.fillna(value=0)
```

`df.fillna(value=0)` replaces the missing values (NaN) in the DataFrame `df` with a specified value (in this case, 0). This operation fills in the NaN values with the given value, creating a new DataFrame with the missing values replaced.

The `df.head()` function is used to display the first few rows of a DataFrame. By default, it returns the first five rows of the DataFrame. This function is helpful for quickly inspecting the structure and content of the DataFrame. It allows you to get a glimpse of the data, including the column names and the values in the top rows. The `head()` function is commonly used at the beginning of data exploration or analysis to get a sense of the dataset's format and to verify if the data is correctly loaded.

```python
df = pd.read_csv('data.csv', header=1)
df
```

The code `df = pd.read_csv('data.csv', header=1)` reads a CSV file named 'data.csv' into a DataFrame object called `df`. It specifies that the second row of the CSV file should be considered as the header row using the `header` parameter.

By using the `read_csv()` function from the pandas library, the CSV file is parsed and converted into a DataFrame. The resulting DataFrame, assigned to the variable `df`, will contain the data from the CSV file, excluding the first row (since it is interpreted as the header row). By setting `header=1`, the DataFrame's column names are taken from the second row of the CSV file, while the actual data starts from the following rows. This is useful when the CSV file has a non-standard or non-existent header row and you want to specify a different row as the header.

```python
df = pd.read_csv('data.csv', nrows=3)
df
```

The code snippet `df = pd.read_csv('data.csv', nrows=3)` reads a CSV file named 'data.csv' into a DataFrame object, but limits the number of rows read to only the first three rows using the `nrows` parameter.

By using the `read_csv()` function from the pandas library, the CSV file is parsed and converted into a DataFrame. The resulting DataFrame, assigned to the variable `df`, will contain the data from the CSV file limited to the specified number of rows (in this case, three). By setting `nrows=3`, only the first three rows of the CSV file will be read into the DataFrame. This can be useful when dealing with large datasets, allowing you to quickly preview a smaller subset of the data without loading the entire file.

```python
df.to_csv('new.csv')
```

The `df.to_csv('new.csv')` command is used to save a DataFrame as a CSV (Comma-Separated Values) file named 'new.csv'. It writes the contents of the DataFrame to the specified file path in CSV format. The resulting CSV file will include the DataFrame's data, column names, and index (unless specified otherwise). This function is useful for exporting DataFrame data to a CSV file, which can be easily shared or used in other applications or analysis tools that support CSV format.
